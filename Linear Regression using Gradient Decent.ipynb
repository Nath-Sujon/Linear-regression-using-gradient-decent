# Steps

```python
m = 0
c = 0
l = 0.0001

n = int(len(x))
iteration = 1000

#loop
for i in range(iteration):
    predy = (m*x) + c
    der_m = (-2/n) * sum(x*(y-predy))
    der_c = (-2/n) * sum(y-predy)
    
    m = (m - ( l * der_m))
    c = (c - (l * der_c))
    
    print(m,c)
```
